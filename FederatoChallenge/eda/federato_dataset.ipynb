```python
# Based on user_session 
columns_to_drop = ['$insert_key', '$schema', 'adid', 'amplitude_attribution_ids', 'amplitude_event_type', 'device_brand', 'device_carrier', 
                   'device_manufacturer', 'device_model', 'global_user_properties', 'idfa', 'is_attribution_event', 'location_lat', 'location_lng', 
                   'partner_id', 'paying', 'sample_rate', 'source_id', 'start_version', 'user_creation_time', 'version_name', 'internalEmailId', 
                   'route', 'path', 'element.props.customAccordion.title', 'option', 'element.key', 'groupId', 'accountId', '[Amplitude] Page Title', 
                   'rules', 'menuVisible', 'filename', '[Amplitude] Page URL', 'hasAccounts', 'variant', 'isCustomAccordion', 'tableId', 'element.props.customAccordion.meta', 
                   'element.props.customAccordion.isPublished', 'quoteName', 'element.props.customAccordion.id', '[Amplitude] Page Domain', 'messageId', 
                   'resourceComponent.featureKey', 'account-id', 'policy-id', 'status', 'referrer', 'formId', 'account-type', 'policyId', 'attachmentId', 
                   'element.props.customAccordion.css', 'error', 'address', 'tab', 'resourceComponent.name', 'udoc', 'menu', 'accountType', '[Amplitude] Page Path', 
                   'buttonType', 'element.props.customAccordion.tab', 'items', 'element.props.customAccordion.__typename', 'item', 'lineOfBusiness', 'line-of-business', 
                   'resourceComponent.featureKeyName', 'element.props.customAccordion.accountType', 'element.props.customAccordion.category', '[Amplitude] Page Counter', 
                   'templateName', 'emailId', 'resourceComponent.ui.defaultVariant', 'element.props.customAccordion.js', 'action', 'name', 'hasAssignees', 
                   'resourceComponent.category', 'referring_domain', 'element.props.name', '[Amplitude] Page Location', 'eventType', 'bazinga', 'data_type', 'data', 
                   'device_type', 'os_version', 'group_properties', 'groups', 'plan', 'platform', 'uuid', '__index_level_0__', 'id', 'version', 'trackingVersion', 
                   'initial_wbraid', 'initial_utm_source', 'initial_msclkid', 'initial_utm_campaign', 'initial_referring_domain', 'initial_utm_id', 'initial_li_fat_id', 
                   'initial_ko_click_id', 'initial_fbclid', 'initial_gclid', 'initial_rtd_cid', 'initial_utm_term','initial_dclid','initial_gbraid','initial_referrer',
                   'initial_utm_medium', 'referring_domain_user', 'initial_utm_content', 'initial_twclid', 'initial_ttclid']
df = df.drop(columns=columns_to_drop)
df.head()
```

```python
df = df[df['user_id'] != 'EMPTY'].reset_index(drop=True)
df.head()
```

```python
# Converting time-related columns to datetime format
timestamp_cols = [
    "client_event_time", "client_upload_time", "event_time", 
    "processed_time", "server_received_time", "server_upload_time"
]
for col in timestamp_cols:
    df[col] = pd.to_datetime(df[col], errors="coerce")

# Group by 'user_id' and 'session_id' to calculate session durations
user_session = df.groupby(['user_id', 'session_id']).agg(
    min_client_event_time=('client_event_time', 'min'),
    max_client_event_time=('client_event_time', 'max')
).reset_index()

# Session time calculation
user_session['session_time'] = user_session['max_client_event_time'] - user_session['min_client_event_time']
user_session['session_seconds'] = user_session['session_time'].dt.total_seconds()

# Merge back into original df
df = df.merge(user_session[['user_id', 'session_id', 'session_seconds']], on=['user_id', 'session_id'], how='left')

df.head()
```

```python
df["time_to_server"] = (df["client_upload_time"] - df["client_event_time"]).dt.total_seconds()
df["server_to_process"] = (df["server_upload_time"] - df["server_received_time"]).dt.total_seconds()
df["processing_time"] = (df["processed_time"] - df["server_upload_time"]).dt.total_seconds()
```

```python
# Function for grouping the events
def categorize_event(event_type):
    if any(keyword in event_type for keyword in ["session_start", "session_end", "application-window", "nav-header", "dashboard"]):
        return "Session & Navigation"
    elif any(keyword in event_type for keyword in ["account", "policy", "rating"]):
        return "Account & Policy Management"
    elif any(keyword in event_type for keyword in ["dashboard", "widget", "layout", "insights", "table"]):
        return "Dashboard & UI Interactions"
    elif any(keyword in event_type for keyword in ["action-center", "task", "workflow", "take-action"]):
        return "Action Center & Workflow"
    elif any(keyword in event_type for keyword in ["submit-click", "form", "create", "definition", "save-click", "submissions"]):
        return "Submission & Forms"
    elif any(keyword in event_type for keyword in ["filter", "sort", "search", "advanced-filters"]):
        return "Filtering & Searching"
    elif any(keyword in event_type for keyword in ["document", "report", "download", "csv"]):
        return "Document & Report Interactions"
    else:
        return "Other/System Events"
```

```python
# Applying function to dataset
df["event_category"] = df["event_type"].apply(categorize_event)
df.head()
```

```python
# Define the mapping for U.S. states to regions
us_region_mapping = {
    'Northeast': ['Connecticut', 'Maine', 'Massachusetts', 'New Hampshire', 'New Jersey', 'New York', 'Pennsylvania', 'Rhode Island', 'Vermont'],
    'Midwest': ['Illinois', 'Indiana', 'Iowa', 'Kansas', 'Michigan', 'Minnesota', 'Missouri', 'Nebraska', 'North Dakota', 'Ohio', 'South Dakota', 'Wisconsin'],
    'South': ['Alabama', 'Arkansas', 'Delaware', 'Florida', 'Georgia', 'Kentucky', 'Louisiana', 'Maryland', 'Mississippi', 'North Carolina', 'Oklahoma', 
              'South Carolina', 'Tennessee', 'Texas', 'Virginia', 'West Virginia'],
    'West': ['Alaska', 'Arizona', 'California', 'Colorado', 'Hawaii', 'Idaho', 'Montana', 'Nevada', 'New Mexico', 'Oregon', 'Utah', 'Washington', 'Wyoming']
}

# Create a function to assign region or 'international'
def assign_region(state):
    # Check if the state is in one of the U.S. regions
    for region, states in us_region_mapping.items():
        if state in states:
            return region
    # If not in U.S. regions, categorize as 'international'
    return 'International'

# Apply function to the dataset
df['region_grouped'] = df['region'].apply(assign_region)
df.head()
```

```python
# Calculate first visit date per user
first_visit = df.groupby("user_id")["event_time"].min().reset_index()
first_visit.columns = ["user_id", "first_visit_date"]

# Merge first visit date back into df
df = df.merge(first_visit, on="user_id", how="left")

# Calculate days since first visit for each event
df["days_since_first_visit"] = (df["event_time"] - df["first_visit_date"]).dt.days

# Find users who returned within 7 days
df["returned_within_7_days"] = df["days_since_first_visit"].between(1, 7).astype(int)

# Merge back to assign a single value per user
user_retention = df[df["days_since_first_visit"] > 0].groupby("user_id")["returned_within_7_days"].max().reset_index()

# Drop the per-event column before merging to avoid overwriting issues
df = df.drop(columns=["returned_within_7_days"], errors="ignore")

# Merge into original df if needed
df = df.merge(user_retention, on="user_id", how="left")

df.head()
```

```python
df['uw'] = df['roles'].str.contains("underwriter").astype(int)
df['admin'] = df['roles'].str.contains("admin").astype(int)
df['manager'] = df['roles'].str.contains("manager").astype(int)
df['broker'] = df['roles'].str.contains("Broker").astype(int)
```

```python
df['google'] = df['referrer_user'].str.contains("google").astype(int)
df['microsoft'] = df['referrer_user'].str.contains("microsoft|teams|office").astype(int)
df['federato'] = df['referrer_user'].str.contains("federato").astype(int)
df['uw-portal'] = df['referrer_user'].str.contains("portal").astype(int)
```

```python
df['client_event_hour'] = df['client_event_time'].dt.hour
df['client_upload_hour'] = df['client_upload_time'].dt.hour
df['event_hour'] = df['event_time'].dt.hour
df['processed_hour'] = df['processed_time'].dt.hour
df['server_received_hour'] = df['server_received_time'].dt.hour
df['server_upload_hour'] = df['server_upload_time'].dt.hour
```

```python
# One hot encoding for appropriate columns
df_encoded = pd.get_dummies(df, columns=["device_family", "event_category", "region_grouped"], dtype=int)
df_encoded.head()
```

```python
final_columns_drop = ['businessUnit', '$insert_id', 'amplitude_id', 'app', 'city', 'device_id', 'event_id', 'event_properties', 'language', 'os_name', 
                      'user_properties', 'rowModel', 'type', '[Amplitude] Session Replay ID', 'first_visit_date', 'days_since_first_visit']

df_encoded = df_encoded.drop(columns=final_columns_drop)
df_encoded.head()
```

```python
# Convert time-related columns to datetime
time_cols = ["event_time", "client_event_time", "processed_time", "server_received_time", 'client_event_time', 'client_upload_time',
             'processed_time','server_received_time', 'server_upload_time']

# Identify one-hot encoded columns (binary columns)
one_hot_cols = [col for col in df_encoded.columns if df_encoded[col].nunique() == 2 and set(df_encoded[col].dropna().unique()) <= {0, 1}]

# Define aggregation rules
agg_funcs = {
    "user_id": "first",  # Assuming one user per session
}

# Min/Max for time columns
for col in time_cols:
    agg_funcs[col] = ["min", "max"]

# One-hot encoded columns: Keep 1 if it appears at least once
for col in one_hot_cols:
    agg_funcs[col] = "max"

# Other numeric columns: Use mean 
numeric_cols = df_encoded.select_dtypes(include=["number"]).columns.difference(one_hot_cols)
for col in numeric_cols:
    agg_funcs[col] = "mean" 

# Aggregate dataset by session_id
df_session = df_encoded.groupby("session_id").agg(agg_funcs).reset_index()

# Flatten multi-index column names
df_session.columns = ["_".join(col).strip("_") for col in df_session.columns]

df_session
```
